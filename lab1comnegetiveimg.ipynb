```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
```

```python
def computeNegative(imge):
    negImge = np.abs(255-imge)
    return negImge
```

```python
url="https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"
imge = io.imread(url)
imge = np.array(imge)
imge = imge.astype('int')
```

```python
plt.imshow(imge,cmap=plt.get_cmap('gray'))
plt.title('orginal image')
plt.show()
```

```python
plt.hist(imge.ravel(),bins = 256, range  = [0,256])
plt.show()
```

```python
color=('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([imge],[i],None,[256],[8,256])
    plt.plot(histr,color = col)
    plt.xlim([8,256])
    plt.show()
```

```python
color=('b','g','r')
for i,col in enumerate(color):
    # Convert the image to uint8 and ensure values are within [0, 255]
    imge_uint8 = imge.astype(np.uint8)
    histr = cv.calcHist([imge_uint8],[i],None,[256],[8,256])
    plt.plot(histr,color = col)
    plt.xlim([8,256])
    plt.show()
```
