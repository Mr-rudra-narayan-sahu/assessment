```python
corpus=[]
for i in range(0,len(dataset)):
    # Indent the lines within the for loop
    review=re.sub('[^a-zA-Z0-9]',' ',dataset['v1,v2,,,,'][i])
    review=review.lower()
    review=review.split() #tokenise
    review=[ps.stem(word) for word in review if not word in stopword_list]
    review=' '.join(review)
    corpus.append(review)
```

```python
 y=pd.get_dummies(dataset['v1,v2,,,,'],drop_first=True)
```

```python
 # Train Test Split
 from sklearn.model_selection import train_test_split
 X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size = 0.20)
 X_train[:3]
 ['lover need',
 'go 4 lunch wif famili aft dat go str 2 orchard lor',
 'pl dont forget studi']
 dataset['v1,v2,,,,'].value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
X_train=cv.fit_transform(X_train).toarray()
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier().fit(X_train,y_train)
```

```python
# Apply CountVectorizer to X_test
X_test = cv.transform(X_test).toarray()

# Now you can predict
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import multilabel_confusion_matrix

# Instead of:
# confusion_matrix(y_test,y_pred)

# Use this:
cm = multilabel_confusion_matrix(y_test, y_pred)

# You can then print the confusion matrix for each class:
for i, matrix in enumerate(cm):
    print(f"Confusion matrix for class {i}:")
    print(matrix)
```
