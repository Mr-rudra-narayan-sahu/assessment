```python
!git clone https://github.com/Mr-rudra-narayan-sahu/assessment.git
!git remote add origin https://github.com/Mr-rudra-narayan-sahu/assessment.git
!git branch -M main
!git push -u origin main
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files



# Convert to Grayscale
def convert_to_grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Thresholding and Binary Pattern
def get_lbp_value(center, neighbors):
    binary = [1 if neighbor >= center else 0 for neighbor in neighbors]
# 8 bit Decimal Conversion
    return sum([b * (2 ** i) for i, b in enumerate(binary)])

# Neighborhood Examination and  Sliding Window
def lbp(image):
    height, width = image.shape
    lbp_image = np.zeros((height, width), dtype=np.uint8)

    for y in range(1, height - 1):
        for x in range(1, width - 1):
            center = image[y, x]
            neighbors = [
                image[y - 1, x - 1], image[y - 1, x], image[y - 1, x + 1],
                image[y, x + 1], image[y + 1, x + 1], image[y + 1, x],
                image[y + 1, x - 1], image[y, x - 1]
            ]
            lbp_value = get_lbp_value(center, neighbors)
            lbp_image[y, x] = lbp_value

    return lbp_image

# Load, preprocess, and apply LBP
image = cv2.imread('/content/assessment/soil.jpg')
grayscale_image = convert_to_grayscale(image)
lbp_image = lbp(grayscale_image)

# Display results
cv2_imshow( image)
cv2_imshow( lbp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
