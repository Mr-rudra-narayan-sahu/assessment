```python
!git clone https://github.com/Mr-rudra-narayan-sahu/assessment.git
!git remote add origin https://github.com/Mr-rudra-narayan-sahu/assessment.git
!git branch -M main
!git push -u origin main
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier

# IRIS dataset
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

# Create color maps
cmap_light = plt.cm.Paired
cmap_bold = ['red', 'c', 'blue']

def plot_decision_boundaries(X, y, classifier, k):
    # Create a mesh grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

    # classifications for each point in mesh grid
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plotting
    plt.figure()
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=cmap_light)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.Paired)
    plt.title(f"k-NN Decision Boundary (k={k},weight='uniform')")
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# k=15
k = 15
weight=('uniform')
classifier = KNeighborsClassifier(n_neighbors=k, weights='uniform')
classifier.fit(X, y)
plot_decision_boundaries(X, y, classifier, k)
# for k=1
k = 1
weight=('uniform')
classifier = KNeighborsClassifier(n_neighbors=k, weights='uniform')
classifier.fit(X, y)
plot_decision_boundaries(X, y, classifier, k)
# for k=5
k = 5
weight=('uniform')
classifier = KNeighborsClassifier(n_neighbors=k, weights='uniform')
classifier.fit(X, y)
plot_decision_boundaries(X, y, classifier, k)


```
